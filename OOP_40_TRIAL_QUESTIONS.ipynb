```python
# Question 15: E-commerce System
class Customer:
    def __init__(self, name, email, balance):
        self.name = name
        self.email = email
        self.balance = balance
        self.purchase_history = []

class EcommerceSite:
    def __init__(self, site_name, available_items):
        self.site_name = site_name
        self.available_items = available_items  # dict with item: quantity pairs

    def purchase_item(self, customer, item, quantity):
        if item not in self.available_items:
            raise ValueError(f"Item {item} not available")
        if self.available_items[item] < quantity:
            raise ValueError(f"Insufficient stock. Only {self.available_items[item]} available")

        total_cost = quantity * 10  # Assuming each item costs 10 units
        if customer.balance < total_cost:
            raise ValueError("Insufficient balance")

        self.available_items[item] -= quantity
        customer.balance -= total_cost
        customer.purchase_history.append((item, quantity, total_cost))

    def refund_item(self, customer, item, quantity):
        # Find the purchase in history
        for purchase in customer.purchase_history:
            if purchase[0] == item and purchase[1] == quantity:
                self.available_items[item] += quantity
                customer.balance += purchase[2]
                customer.purchase_history.remove(purchase)
                return
        raise ValueError("No matching purchase found for refund")

    def display_purchase_history(self, customer):
        if not customer.purchase_history:
            return "No purchase history"
        return "\n".join([f"Item: {item}, Quantity: {qty}, Cost: {cost}"
                         for item, qty, cost in customer.purchase_history])

# Question 16: Sports Competition System
class Athlete:
    def __init__(self, name, sport, ranking):
        self.name = name
        self.sport = sport
        self.ranking = ranking

class Competition:
    def __init__(self, competition_name, date):
        self.competition_name = competition_name
        self.date = date
        self.athletes = []

    def register_athlete(self, athlete):
        if athlete not in self.athletes:
            self.athletes.append(athlete)
        else:
            raise ValueError("Athlete already registered")

    def remove_athlete(self, athlete):
        if athlete in self.athletes:
            self.athletes.remove(athlete)
        else:
            raise ValueError("Athlete not registered in competition")

    def display_competition_details(self):
        details = f"Competition: {self.competition_name}\nDate: {self.date}\nAthletes:"
        for athlete in self.athletes:
            details += f"\n- {athlete.name} ({athlete.sport}, Rank: {athlete.ranking})"
        return details

# Question 17: Company Management System
class Company:
    def __init__(self, company_name, industry, total_employees):
        self.company_name = company_name
        self.industry = industry
        self.total_employees = total_employees

class Department:
    def __init__(self, department_name, manager_name, employee_count):
        self.department_name = department_name
        self.manager_name = manager_name
        self.employee_count = employee_count
        self.employees = []

    def add_employee(self, employee_name):
        if employee_name not in self.employees:
            self.employees.append(employee_name)
            self.employee_count += 1
        else:
            raise ValueError("Employee already in department")

    def remove_employee(self, employee_name):
        if employee_name in self.employees:
            self.employees.remove(employee_name)
            self.employee_count -= 1
        else:
            raise ValueError("Employee not found in department")

    def display_department_info(self):
        return f"Department: {self.department_name}\nManager: {self.manager_name}\n" \
               f"Employee Count: {self.employee_count}\nEmployees: {', '.join(self.employees)}"

# Question 18: Inventory Management System
class Product:
    def __init__(self, product_name, product_code, price):
        self.product_name = product_name
        self.product_code = product_code
        self.price = price

class Inventory:
    def __init__(self, inventory_id):
        self.inventory_id = inventory_id
        self.product_list = {}  # product_name: quantity

    def add_product(self, product_name, quantity):
        if product_name in self.product_list:
            self.product_list[product_name] += quantity
        else:
            self.product_list[product_name] = quantity

    def remove_product(self, product_name, quantity):
        if product_name not in self.product_list:
            raise ValueError("Product not in inventory")
        if self.product_list[product_name] < quantity:
            raise ValueError("Insufficient quantity")

        self.product_list[product_name] -= quantity
        if self.product_list[product_name] == 0:
            del self.product_list[product_name]

    def display_inventory(self):
        return "\n".join([f"{product}: {quantity}"
                         for product, quantity in self.product_list.items()])

# Question 19: Hotel Reservation System
class Hotel:
    def __init__(self, hotel_name, location, rooms_available):
        self.hotel_name = hotel_name
        self.location = location
        self.rooms_available = rooms_available
        self.reserved_rooms = {}  # room_number: guest_name

class Reservation:
    def __init__(self, guest_name, check_in_date, room_number):
        self.guest_name = guest_name
        self.check_in_date = check_in_date
        self.room_number = room_number

    def book_room(self, hotel):
        if self.room_number in hotel.reserved_rooms:
            raise ValueError("Room already booked")
        hotel.reserved_rooms[self.room_number] = self.guest_name
        hotel.rooms_available -= 1

    def cancel_reservation(self, hotel):
        if self.room_number not in hotel.reserved_rooms:
            raise ValueError("Reservation not found")
        if hotel.reserved_rooms[self.room_number] != self.guest_name:
            raise ValueError("Reservation doesn't match guest")

        del hotel.reserved_rooms[self.room_number]
        hotel.rooms_available += 1

    def display_reservation_details(self):
        return f"Guest: {self.guest_name}\nCheck-in Date: {self.check_in_date}\n" \
               f"Room Number: {self.room_number}"

# Question 20: Banking System
class BankCustomer:
    def __init__(self, name, email, phone):
        self.name = name
        self.email = email
        self.phone = phone

class Account:
    def __init__(self, account_number, balance, account_type):
        self.account_number = account_number
        self.balance = balance
        self.account_type = account_type

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.balance += amount
        return f"Deposited ${amount}. New balance: ${self.balance}"

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.balance:
            raise ValueError("Insufficient funds")
        self.balance -= amount
        return f"Withdrawn ${amount}. New balance: ${self.balance}"

    def display_account_balance(self):
        return f"Account Number: {self.account_number}\nType: {self.account_type}\n" \
               f"Current Balance: ${self.balance}"
```

```python
from datetime import datetime
import re

# Question 21: Article Publishing System
class Author:
    def __init__(self, name, nationality, birth_year):
        self.name = name
        self.nationality = nationality
        self.birth_year = birth_year

class Article(Author):
    def __init__(self, name, nationality, birth_year):
        super().__init__(name, nationality, birth_year)
        self.title = None
        self.published_year = None

    def publish_article(self, title, year):
        current_year = datetime.now().year
        if year > current_year:
            raise ValueError(f"Cannot publish in future year {year}")
        self.title = title
        self.published_year = year

    def display_article_info(self):
        if not self.title:
            return "Article not yet published"
        return f"Title: {self.title}\nAuthor: {self.name}\nPublication Year: {self.published_year}"

# Question 22: User Management System
class User:
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self._password = password  # Using protected attribute for password

class Admin(User):
    def __init__(self, username, email, password):
        super().__init__(username, email, password)
        self.role = 'Admin'
        self.user_database = {}  # Store user accounts

    def create_user(self, username, email, password):
        if username in self.user_database:
            raise ValueError(f"Username {username} already exists")
        if not self._validate_email(email):
            raise ValueError("Invalid email format")
        self.user_database[username] = User(username, email, password)

    def delete_user(self, username):
        if username not in self.user_database:
            raise ValueError(f"User {username} does not exist")
        del self.user_database[username]

    def display_user_info(self, username):
        if username not in self.user_database:
            raise ValueError(f"User {username} not found")
        user = self.user_database[username]
        return f"Username: {user.username}\nEmail: {user.email}\nRole: User"

    @staticmethod
    def _validate_email(email):
        pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
        return re.match(pattern, email) is not None

# Question 23: Device Management System
class Device:
    def __init__(self, device_name, model, status="off"):
        self.device_name = device_name
        self.model = model
        self.status = status

class Smartphone(Device):
    def __init__(self, device_name, model, battery_level, storage_capacity):
        super().__init__(device_name, model)
        self.battery_level = battery_level
        self.storage_capacity = storage_capacity

    def turn_on(self):
        if self.status == "on":
            raise ValueError("Device is already turned on")
        if self.battery_level <= 0:
            raise ValueError("Cannot turn on: Battery is depleted")
        self.status = "on"

    def turn_off(self):
        if self.status == "off":
            raise ValueError("Device is already turned off")
        self.status = "off"

    def display_device_info(self):
        return f"Device: {self.device_name}\nModel: {self.model}\nStatus: {self.status}\n" \
               f"Battery: {self.battery_level}%\nStorage: {self.storage_capacity}GB"

# Question 24: Classroom Management System
class Teacher:
    def __init__(self, name, subject, years_of_experience):
        self.name = name
        self.subject = subject
        self.years_of_experience = years_of_experience
        self.assigned_classroom = None

class Classroom:
    def __init__(self, class_name, capacity):
        self.class_name = class_name
        self.capacity = capacity
        self.assigned_teacher = None

    def assign_teacher(self, teacher):
        if teacher.assigned_classroom:
            raise ValueError(f"Teacher {teacher.name} is already assigned to another classroom")
        self.assigned_teacher = teacher
        teacher.assigned_classroom = self

    def display_classroom_info(self):
        teacher_info = f"Teacher: {self.assigned_teacher.name}" if self.assigned_teacher else "No teacher assigned"
        return f"Classroom: {self.class_name}\nCapacity: {self.capacity}\n{teacher_info}"

# Question 25: Music Management System
class Artist:
    def __init__(self, name, genre, number_of_albums):
        self.name = name
        self.genre = genre
        self.number_of_albums = number_of_albums

class Song(Artist):
    def __init__(self, name, genre, number_of_albums):
        super().__init__(name, genre, number_of_albums)
        self.title = None
        self.duration = None

    def add_song(self, title, duration=None):
        if duration is None:
            raise ValueError("Duration must be specified")
        if duration <= 0:
            raise ValueError("Duration must be positive")
        self.title = title
        self.duration = duration

    def display_song_info(self):
        if not self.title:
            return "Song not yet added"
        return f"Title: {self.title}\nArtist: {self.name}\nDuration: {self.duration} seconds"

# Question 26: Course Management System
class Course:
    def __init__(self, course_name, course_code, credits):
        self.course_name = course_name
        self.course_code = course_code
        self.credits = credits

class Instructor:
    def __init__(self, name, department):
        self.name = name
        self.department = department
        self.courses_taught = []

    def assign_course(self, course):
        if course in self.courses_taught:
            raise ValueError(f"Course {course.course_name} is already assigned to this instructor")
        self.courses_taught.append(course)

    def display_instructor_info(self):
        courses = "\n".join([f"- {course.course_name} ({course.course_code})"
                           for course in self.courses_taught])
        return f"Instructor: {self.name}\nDepartment: {self.department}\nCourses:\n{courses}"

# Question 27: Order Management System
class Product:
    def __init__(self, product_name, price, stock):
        self.product_name = product_name
        self.price = price
        self.stock = stock

class Order:
    def __init__(self, order_number):
        self.order_number = order_number
        self.ordered_products = []

    def add_product(self, product):
        if product.stock <= 0:
            raise ValueError(f"Product {product.product_name} is out of stock")
        self.ordered_products.append(product)
        product.stock -= 1

    def remove_product(self, product):
        if product not in self.ordered_products:
            raise ValueError(f"Product {product.product_name} not in order")
        self.ordered_products.remove(product)
        product.stock += 1

    def display_order_info(self):
        products = "\n".join([f"- {product.product_name} (${product.price})"
                            for product in self.ordered_products])
        return f"Order #{self.order_number}\nProducts:\n{products}"

# Question 28: Customer Purchase System
class Customer:
    def __init__(self, name, email, loyalty_points=0):
        self.name = name
        self.email = email
        self.loyalty_points = loyalty_points

class Purchase(Customer):
    def __init__(self, name, email):
        super().__init__(name, email)
        self.purchases = []

    def make_purchase(self, amount, use_points=False):
        if use_points:
            points_value = self.loyalty_points * 0.01  # 1 point = $0.01
            if points_value < amount:
                raise ValueError("Insufficient loyalty points")
            self.loyalty_points -= amount * 100
        self.purchases.append(amount)
        self.add_loyalty_points(int(amount))

    def add_loyalty_points(self, points):
        self.loyalty_points += points

    def display_customer_info(self):
        total_spent = sum(self.purchases)
        return f"Customer: {self.name}\nEmail: {self.email}\n" \
               f"Loyalty Points: {self.loyalty_points}\nTotal Spent: ${total_spent}"

# Question 29: Gadget Management System
class Gadget:
    def __init__(self, gadget_name, brand, price):
        self.gadget_name = gadget_name
        self.brand = brand
        self.price = price

class Laptop(Gadget):
    def __init__(self, gadget_name, brand, price, screen_size, ram):
        super().__init__(gadget_name, brand, price)
        self.screen_size = screen_size
        self.ram = ram

    def upgrade_ram(self, new_size):
        if new_size <= self.ram:
            raise ValueError(f"New RAM size must be greater than current size ({self.ram}GB)")
        old_ram = self.ram
        self.ram = new_size
        return f"RAM upgraded from {old_ram}GB to {new_size}GB"

    def display_gadget_info(self):
        return f"Laptop: {self.gadget_name}\nBrand: {self.brand}\nPrice: ${self.price}\n" \
               f"Screen Size: {self.screen_size} inches\nRAM: {self.ram}GB"

# Question 30: Real Estate Management System
class House:
    def __init__(self, address, num_rooms, price):
        self.address = address
        self.num_rooms = num_rooms
        self.price = price

class RealEstateAgent:
    def __init__(self, name, agency):
        self.name = name
        self.agency = agency
        self.listed_houses = []

    def list_house(self, house):
        if house in self.listed_houses:
            raise ValueError("House is already listed")
        self.listed_houses.append(house)

    def remove_house(self, house):
        if house not in self.listed_houses:
            raise ValueError("House is not listed with this agent")
        self.listed_houses.remove(house)

    def display_agent_info(self):
        houses = "\n".join([f"- {house.address} ({house.num_rooms} rooms, ${house.price})"
                          for house in self.listed_houses])
        return f"Agent: {self.name}\nAgency: {self.agency}\nListed Houses:\n{houses}"
```

```python
# Sorting Algorithms Implementation

def bubble_sort(arr):
    """Bubble Sort Algorithm"""
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap
    return arr

def selection_sort(arr):
    """Selection Sort Algorithm"""
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]  # Swap
    return arr

def insertion_sort(arr):
    """Insertion Sort Algorithm"""
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def merge_sort(arr):
    """Merge Sort Algorithm"""
    if len(arr) > 1:
        mid = len(arr) // 2  # Finding the mid of the array
        L = arr[:mid]  # Dividing the elements into 2 halves
        R = arr[mid:]

        merge_sort(L)  # Sorting the first half
        merge_sort(R)  # Sorting the second half

        i = j = k = 0

        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

def quick_sort(arr):
    """Quick Sort Algorithm"""
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)

def heapify(arr, n, i):
    """Helper function to maintain heap property"""
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Swap
        heapify(arr, n, largest)

def heap_sort(arr):
    """Heap Sort Algorithm"""
    n = len(arr)

    # Build a maxheap.
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # One by one extract elements from heap.
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Swap
        heapify(arr, i, 0)

    return arr

# Example usage of sorting algorithms
if __name__ == "__main__":
    unsorted_list = [64, 34, 25, 12, 22, 11, 90]

    print("Original list:", unsorted_list)

    print("Bubble Sorted:", bubble_sort(unsorted_list.copy()))
    print("Selection Sorted:", selection_sort(unsorted_list.copy()))
    print("Insertion Sorted:", insertion_sort(unsorted_list.copy()))

    print("Merge Sorted:", merge_sort(unsorted_list.copy()))

    print("Quick Sorted:", quick_sort(unsorted_list.copy()))

    print("Heap Sorted:", heap_sort(unsorted_list.copy()))
```
